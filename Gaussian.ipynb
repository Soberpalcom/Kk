```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()

# Convert data to a DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Standardize the data
iris_df_scaled = (iris_df - iris_df.mean()) / iris_df.std()
```

```python
# Reduce dimensionality with PCA
pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_df_scaled)
```

```python
# Initialize and fit Gaussian Mixture model
gmm = GaussianMixture(n_components=3)
gmm.fit(iris_pca)

# Predict cluster labels
cluster_labels = gmm.predict(iris_pca)
```

```python
plt.figure(figsize=(8, 6))

# Plot data points colored by cluster label
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c=cluster_labels, cmap='viridis')

# Add centroids=                                                                                                                                                                                                                                                                                                                                                                                                                                  o
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='black', marker='x', s=200, label='Centroids')

# Add labels and title
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clustering the Iris Dataset using Gaussian Mixture Models')
plt.legend()
plt.show()
```
